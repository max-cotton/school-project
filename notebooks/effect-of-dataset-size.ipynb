```python
import os

import matplotlib.pyplot as plt
import numpy as np

from school_project.models.gpu.mnist import MNISTModel as Model

os.chdir(os.getcwd())

train_dataset_sizes = np.array(list(range(1, 209, 10)))
test_prediction_accuracies = np.array([])

print(f"Train Dataset Sizes: {train_dataset_sizes}")

for index, train_dataset_size in enumerate(train_dataset_sizes):

    print(f"Train Dataset Size: {train_dataset_size}, {round(number=index/len(train_dataset_sizes) * 100, ndigits=2)}% of the way done.")

    model = Model(
        hidden_layers_shape = [100, 100],
        train_dataset_size = train_dataset_size,
        learning_rate = 0.1,
        use_relu = True
        )
    model.create_model_values()

    model.train(100)

    model.test()

    test_prediction_accuracies = np.append(test_prediction_accuracies, model.test_prediction_accuracy)

m, c = np.polyfit(train_dataset_sizes, test_prediction_accuracies, 1)

plt.xlabel(xlabel="Train Dataset Size")
plt.ylabel(ylabel="Test Prediction Accuracy")
plt.scatter(train_dataset_sizes, test_prediction_accuracies, marker='x')
plt.plot(train_dataset_sizes, m*train_dataset_sizes + c)
plt.show()
```
